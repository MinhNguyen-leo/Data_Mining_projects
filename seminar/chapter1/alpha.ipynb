## Additionally

### Import the libraries

```python
import requests
import pandas as pd
from io import BytesIO
```

### Download Bitcoin’s intraday data

```python
AV_API_URL = "https://www.alphavantage.co/query"
parameters = {
"function": "CRYPTO_INTRADAY",
"symbol": "ETH",
"market": "USD",
 "interval": "30min",
    "outputsize": "full",
    "apikey": ALPHA_VANTAGE_API_KEY
}
r = requests.get(AV_API_URL, params=parameters)
data = r.json()
df = (
    pd.DataFrame(data["Time Series Crypto (30min)"])
    .transpose()
)
df
```

### Download the upcoming earnings announcements within the next three months

```python
AV_API_URL = "https://www.alphavantage.co/query"
parameters = {
    "function": "EARNINGS_CALENDAR",
    "horizon": "3month",
    "apikey": ALPHA_VANTAGE_API_KEY
}
r = requests.get(AV_API_URL, params=parameters)
pd.read_csv(BytesIO(r.content))
```
