### Import libraries

```python
import pandas as pd
import numpy as np
import yfinance as yf
```

### Download data

```python
df = yf.download("AAPL",
                 start="2010-01-01",
                 end="2020-12-31",
                 progress=False)
df = df.loc[:, ["Adj Close"]]
```

### Calculating the simple and log returns using the adjusted close prices

df["simple_rtn"] = df["Adj Close"].pct_change()
df["log_rtn"] = np.log(df["Adj Close"]/df["Adj Close"].shift(1))
